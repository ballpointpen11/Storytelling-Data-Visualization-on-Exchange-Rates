# Storytelling Data Visualization on Exchange Rates
The dataset we're using describes Euro daily exchange rates between 1999 and 2021.  

If the exchange rate of the euro to the US dollar is 1.5, you get 1.5 US dollars if you pay 1.0 euro (one euro has more value than one US dollar at this exchange rate). 

### Importing and Reading the Dataset

```python
# Importing Pandas to use on the dataset.
import pandas as pd
# Reading in the data set and assign it to the variable: exchange_rates.
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv', encoding="utf8")
```

This dataset has 5,878 rows and 45 columns.  There are not any null values but most of the dtypes are objects and not floats.

## Data Cleaning
Below, we do the following:

- We rename the [US dollar ] and Period\Unit: columns to something easier to type — US_dollar and Time.
- We change the Time column to a datetime data type.
- We sort the values by Time in ascending order.
- We reset the index (and drop the initial index).


```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                               inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
# Isolating the 'Time' and 'US_dollar' columns to 'euro_to_dollar'.
euro_to_dollar = exchange_rates[['Time','US_dollar']]
print(euro_to_dollar['US_dollar'].value_counts())
```

When checking the value counts of 'US_dollar', the dash (-) has 62 counts while the next highest value has a count of 9.  We need to drop the dash (-) so we can convert the 'US_dollar' column to a float dtype.

```python
# Removing the dash (-) from the 'US_dollar' column.
euro_to_dollar = euro_to_dollar.loc[euro_to_dollar['US_dollar'] != '-']
```

```python
# Converting the 'US_dollar' column to the float dtype.
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

```python
# Checking the 'US_dollar' value counts again.
euro_to_dollar.value_counts()
```

## Rolling Mean
The data cleaning is done, now we will generate a line plot to visualize the evolution of the euro-dollar exchange rate.

```python
import matplotlib.pyplot as plt
#Enables Jupyter to display graphs
%matplotlib inline 


plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

If we look at the line's shape, we see many small wiggles — rather than seeing a smooth line.  Below we are extending the rolling window to 30 days.  This means that each point is the mean of the last 30 days.  This creates a smoother line

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
```

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['rolling_mean'])
plt.show()
```

## Creating a Storytelling Data Visualization

Below a grid chart of four line plots will be created to visualize the Euro to US dollar exchange rate in increments of five years.
https://nbviewer.org/urls/community.dataquest.io/uploads/short-url/cE2nD4ZwPr5pWPQNXkY2kkilPOn.ipynb Delete when done

```python
import datetime
import matplotlib.pyplot as plt
import matplotlib.dates 
import matplotlib.ticker as mticker
```

```python
# Configuring 4 subplots, 1 row and 4 columns.
fix, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=1, ncols=4,
                                        figsize=(12,5))
axes = [ax1, ax2, ax3, ax4]

# Removing the top and right 'spines' of each subplot.
for ax in axes:
    for location in ['top','right','left','bottom']:
        ax.spines[location].set_visible(False)
        ax.xaxis.set_major_locator(plt.LinearLocator(5))
        ax.yaxis.set_major_locator(plt.LinearLocator(4))

# Adding the year title at the top left of each subplot.
ax.text(-3.3, 1.05,'2001-2005',size=12,
     horizontalalignment='center',
     verticalalignment='center',
     transform = ax.transAxes)
ax.text(-2.1, 1.05,'2006-2010',size=12,
     horizontalalignment='center',
     verticalalignment='center',
     transform = ax.transAxes)
ax.text(-0.9, 1.05,'2011-2015',size=12,
     horizontalalignment='center',
     verticalalignment='center',
     transform = ax.transAxes)
ax.text(0.3, 1.05,'2016-2020',size=12,
     horizontalalignment='center',
     verticalalignment='center',
     transform = ax.transAxes)

# First subplot.
ax1.plot(euro_to_dollar['Time'],
         euro_to_dollar['rolling_mean'])
ax1.set_xlim([datetime.date(2000, 12, 31),datetime.date(2006, 1, 1)])
ax1.set_yticklabels(["1.0","1.2","1.4","1.6"])
ax1.set_xticklabels(["'01","'02","'03","'04","'05"],rotation=45)
ax1.axhline(1.2029966666666665,color='grey',alpha=0.75,
           xmin=0.35,xmax=1.1,linestyle='--')

# Second subplot.
ax2.plot(euro_to_dollar['Time'],
         euro_to_dollar['rolling_mean'])
ax2.set_xlim([datetime.date(2006, 1, 1),datetime.date(2010, 1, 1)])
ax2.set_yticklabels(["1.0","1.2","1.4","1.6"])
ax2.set_xticklabels(["'06","'07","'08","'09","'10"],rotation=45)
ax2.axhline(1.2029966666666665,color='grey',alpha=0.75,
           xmin=0,xmax=1,linestyle='--')

# Third subplot.
ax3.plot(euro_to_dollar['Time'],
         euro_to_dollar['rolling_mean'])
ax3.set_xlim([datetime.date(2010, 1, 1),datetime.date(2015, 12, 31)])
ax3.set_yticklabels(["1.0","1.2","1.4","1.6"])
ax3.set_xticklabels(["'11","'12","'13","'14","'15"],rotation=45)
ax3.axhline(1.2029966666666665,color='grey',alpha=0.75,
           xmin=0.05,xmax=1,linestyle='--')

# Fourth subplot.
ax4.plot(euro_to_dollar['Time'],
         euro_to_dollar['rolling_mean'])
ax4.set_xlim([datetime.date(2016, 1, 1),datetime.date(2020, 12, 31)])
ax4.set_yticklabels(["1.0","1.2","1.4","1.6"])
ax4.set_xticklabels(["'16","'17","'18","'19","'20"],rotation=45)
ax4.axhline(1.2029966666666665,color='grey',alpha=0.75,
           xmin=0,xmax=1,linestyle='--')

plt.show()
```

The above graph visualizes the Euro to dollar exchange rate in four seperate subplots.  Plot one is from 2001-2005, plot two is from 2006-2010, plot three is from 2011-2015, and plot four is from 2016-2020.  The dashed line is the median exchange rate of the rolling mean of 30 days we calculated, which is 1.202.

The median line in this graph shows how the 2007-08 financial crisis fluctuated the exchange rate and how long it took to recover to the median line.  It took about 9 years for the exchange rate to get below the median line.  As the exchange rate increases, the US dollar is weaker against the Euro.
